# **METODOS DE LISTA**

# **PERCORRENDO E FATIANDO STRINGS**

```python
# PERCORRENDO LISTA COM FOR

cars = ["bmw", "audi", "porshe", "ferrari"]
print(cars)

for x in cars:
    print(x)
```

```python
# PERCORRENDO LISTA USANDO RANGE DE LISTA
for x in range(len(cars)):
  print(cars[x])
```

```python
# usando while

contador = 0
while contador < len(cars):
  print(cars[contador])
  contador += 1
```

```python
#FATIANDO LISTAS
compras = ["arroz", "feijão", "farinha de trigo", "mainteiga"]
```

```python
print(compras[0:5:3])
print(compras[::])
print(compras[-1:-3:-1])
```

# **TUPLAS**

uma tupla, é uma estrutura de dados utilizada também para armazenar mais de um dado por vez na mesma estrutura. Porem ela tem caracteristicas difernetes de uma listas. ( Tuplas são imutaveis)

```python
# tuplas e como alterar dados

frutas = ("banana", "maça", "abacaxi", "melancia")


# não consigo alterar o valor de uma tupla diretamente,
#então converto seu tipo para lista, e posso usar todos os demais metodos


auxiliar = list(frutas)

# e depois posso converter para tupla dnv!

frutas = tuple(auxiliar)
```

```python
# extrair dados de tupla para variavel(atribuição)
clubes = ("galo", "barcelona", "chelsea")

(n1, n2, n3) = clubes

print(f"Meu clube do coração é o {n1}, meu segundo favorito é o {n2}, e só depois vem o maior de londres, o {n3}!")

numeros = (1,2,2,2,2,2,2,3,3,3,3,33,2,4)
(n1, n2, *n3) = numeros

for i in range(len(numeros)):
    print(i)

```

```python
# juntar tuplas

frutas1 = ("morango", "morango")
frutas2 = ("banana", "maça")
frutas3 = frutas1 + frutas2
print(frutas3)
```

```python
#count

frutas3.count("morango")
```

```python
# index
frutas3.index("maça")
```

#**LiStas Aninhadas**

```python
# Lista aninhada (encadeada)

colaboradores = [
    ["Nome", "Cargo", "Departamento"],
    ["Morgan", "Analista", "Financeiro"],
    ["Edward", "Maneger", "Vendas"],
    ["Marks", "CEO", "Liderança"]
]
```

```python
# acessando dados

print(colaboradores[3][1])
```

```python
# modificando um elemento da lista aninhada
colaboradores[1][1] = "CFO"
print(colaboradores[1][1])
```

```python
# percorrer lista aninhada

for lista in colaboradores:
    for dados in lista:
      print(dados, end=" ")
    print()
```

```python

#extraindo dados de uma lista e jogando em uma variavel

for dados in colaboradores[1:]:
  nome, cargo, departamento = dados
  print (f"{nome},  {cargo}, {departamento}")
```

```python
# exemplo com range

for i in range(len(colaboradores)):
    for j in range(len(colaboradores[i])):
      print(colaboradores[i][j])
    print()
```

# **SETS**

CONJUSTOS (SETS) é uma coleção não ordenada, mutavel e qie mão permite elementos duplicados

```python
# sintaxe de um SET

frutas = {"banana", "maçã", "uva", "Melância", "limão"}
```

```python
# definir set vazio

frutis = set()
print(type(frutis))
```

```python
# tamanho de um set

print(len(frutas))

```

```python
# valores duplicados não são aceitos, se vocÊ usar, o set o ignora
# ele tambem entrega uma ordem aleatorio, não tem indice

```

```python
#Acessando itens no set
for x in frutas:
  print(x)
```

```python
numeros = {1,2,3,4,1,4,5,6,6,6,7}

for x in numeros:
  print(x)
```

```python
# verificando se um item existe no set

if "maçã" in frutas:
  print("tem")
else:
  print("Não tem")
```

```python
"""não é possivel editar um elemento,
mas é possivel adicionar novos elementos
"""
frutas.add("morango")
print(frutas)

"""
a funçaõ update tambem adicona novos elemntos, mas ele intera
cada item, tornando uma string ou qualquer outro elemento
em item individual!!!

"""
```

```python
#concatenando Sets

cor1 = {"preto "}
cor2 = {" branco"}

cor1.update(cor2)
print(cor1)
```

```python
#excluindo valores do set

frutas.remove("morango")
print(frutas)
```

```python
# verificar se Duas listas são iguais

carros = ["bmw", "ford", "fiat"]
carros2 = ["bmw", "ford", "fiat", "fiat"]

# a lista não é igual, mas c vc transformar em set
# elas vão ser nesse caso, pois vai retirar os valores repetidos


if set(carros) == set(carros2):
    print("São iguais")
else:
    print("Não são iguais")
```

```python
# metodo copy

aux = a.copy()
print(aux)
```

```python
# metodo de união ( juntar elementos dos conjuntos)

#1 forma

aux = a.union(b)
print(aux)

#2 forma
auxi = a | b
print(auxi)
```

```python
# metodo de interserção ( apenas os elementos em comum entre os conjuntos)
a = {1, 2, 3}
b = {3, 5, 1}

#1 forma

aux1 = a.intersection(b)
aux2 = a & b

print(aux2)
```

```python
#diferença (apenas o que tem no primeiro conjunto e não tem no segundo)
x = {1, 2, 3, 9}
z = {3, 5, 1, 6}

aux = x.difference(z)
aux1 = c - b
print(aux)
```

```python

# diferença simetrica ( oque tem no primeiro e segundo conjunto, mas nao no resto)
x = {1, 2, 3, 9}
z = {3, 5, 1, 6}

aux = x.symmetric_difference(z)
print(aux)
```
